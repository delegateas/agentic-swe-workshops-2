## 1.4 Øvelse 1: Forstå Execute-funktionerne til ChargeSmart Tools
*~15 minutter*

Nu skal du **forstå og evt. modificere** `execute`-funktionerne for 3 ChargeSmart tools.

**📋 Opgave:** Før du kører cellen nedenfor:
1. Læs koden igennem og forstå hvad hver funktion gør
2. Prøv at skjule koden og skrive din egen version (brug hints nedenfor)
3. Sammenlign med løsningen og kør cellen

**Du har adgang til:**
- `stations` - Map med alle stationer (ID → StationInfo)
- `activeSessions` - Map med aktive sessioner (stationId → session info)
- `getTariff(hour)` - Returnerer pris pr. kWh for en given time

**Krav til de 3 tools:**

| Tool | Input | Output |
|------|-------|--------|
| `get_station_status` | stationId (string) | Status på én station (ledig/optaget, kW, type) |
| `list_stations` | *ingen* | Liste over alle stationer med status |
| `calculate_charging_cost` | kwh (number), hour (int) | Pris for at lade X kWh på tidspunkt Y |

````typescript
// 🎯 ØVELSE 1: Studer disse Execute-funktioner!
// Tip: args er et Record<string, unknown> - brug args.stationId as string osv.

const tools = new Map<string, SimpleTool>([
  ["get_station_status", {
    name: "get_station_status",
    description: "Get real-time status of an EV charging station including availability, power level, and current session info",
    parameterSchema: {
      type: "object",
      properties: { stationId: { type: "string", description: "Station ID like CPH-001" } },
      required: ["stationId"],
    },
    execute: (args) => {
      // 1. Hent stationId fra args
      const stationId = args.stationId as string;

      // 2. Slå op i stations map
      const station = stations.get(stationId);
      if (!station) return `Error: Station ${stationId} not found`;

      // 3. Tjek om der er en aktiv session
      const session = activeSessions.get(stationId);

      // 4. Returner beskrivende streng
      if (session) {
        return `Station ${stationId} (${station.name}): IN USE by ${session.vehicleId} since ${session.start.toTimeString().slice(0, 5)}, ${session.kwh.toFixed(1)} kWh delivered. ${station.powerKw}kW ${station.type} charger.`;
      }
      return `Station ${stationId} (${station.name}): AVAILABLE. ${station.powerKw}kW ${station.type} charger at ${station.location}.`;
    },
  }],
  ["list_stations", {
    name: "list_stations",
    description: "List all EV charging stations in the ChargeSmart network with their current availability",
    parameterSchema: { type: "object", properties: {} },
    execute: (_args) => {
      const lines = [...stations.values()].map(s => {
        const status = activeSessions.has(s.id) ? "IN USE" : "AVAILABLE";
        return `- ${s.id}: ${s.name} (${s.location}) - ${s.powerKw}kW ${s.type} - ${status}`;
      });
      return "ChargeSmart Network Stations:\n" + lines.join("\n");
    },
  }],
  ["calculate_charging_cost", {
    name: "calculate_charging_cost",
    description: "Calculate the cost of charging based on kWh needed and time of day",
    parameterSchema: {
      type: "object",
      properties: {
        kwh: { type: "number", description: "Energy needed in kWh" },
        hour: { type: "integer", description: "Hour of day (0-23)" },
      },
      required: ["kwh", "hour"],
    },
    execute: (args) => {
      const kwh = Number(args.kwh);
      const hour = Number(args.hour);
      const tariff = getTariff(hour);
      const cost = kwh * tariff;
      const period = (hour >= 0 && hour < 6) ? "off-peak" : (hour >= 17 && hour < 21) ? "peak" : "normal";
      return `Charging ${kwh.toFixed(1)} kWh at ${hour.toString().padStart(2, '0')}:00 (${period} tariff: ${tariff.toFixed(2)} DKK/kWh) = ${cost.toFixed(2)} DKK`;
    },
  }],
]);

console.log(`✅ Defineret ${tools.size} tools:`);
for (const tool of tools.values()) {
  console.log(`   - ${tool.name}: ${tool.description.substring(0, 50)}...`);
}
````

## 1.5 The Agentic Loop - ~50 Lines - det er det hele

Nu bygger vi det komplette agentic loop fra bunden. Vi skal blot først oversætte vores SimpleTool til Ollama's tool-format.

````typescript
import ollama from "npm:ollama";

console.log("✅ ollama pakke loaded!");
````

````typescript
// Hjælpefunktion: Konverter SimpleTool til Ollama's tool format
function toOllamaFormat(tool: SimpleTool) {
  return {
    type: "function" as const,
    function: {
      name: tool.name,
      description: tool.description,
      parameters: tool.parameterSchema,
    },
  };
}

// THE AGENTIC LOOP - ~50 lines, no magic!
async function runAgent(userMessage: string, tools: Map<string, SimpleTool>): Promise<string> {
  const messages: { role: string; content: string }[] = [
    {
      role: "system",
      content: `Du er en hjælpsom EV-ladnings assistent for ChargeSmart netværket i København.
        Brug de tilgængelige tools til at besvare spørgsmål om ladestationer,
        tilgængelighed, og priser. Svar altid på dansk.`,
    },
    { role: "user", content: userMessage },
  ];

  const ollamaTools = [...tools.values()].map(toOllamaFormat);
  const maxIterations = 10;

  for (let iteration = 1; iteration <= maxIterations; iteration++) {
    console.log(`\n--- Iteration ${iteration} ---`);

    // 1. Send til Ollama med tool-definitioner
    const response = await ollama.chat({
      model: selectedModelName,
      messages,
      tools: ollamaTools,
    });

    messages.push(response.message as any);

    // 2. Tjek for tool calls
    if (!response.message.tool_calls?.length) {
      // Ingen tool calls = LLM er færdig
      console.log("✅ LLM finished (no tool calls)");
      return response.message.content || "No response";
    }

    // 3. Eksekvér hver tool call SELV
    console.log(`🔧 LLM wants to call ${response.message.tool_calls.length} tool(s)`);

    for (const toolCall of response.message.tool_calls) {
      const toolName = toolCall.function.name;
      const toolArgs = toolCall.function.arguments as Record<string, unknown>;
      console.log(`   Tool: ${toolName}`);
      console.log(`   Args: ${JSON.stringify(toolArgs)}`);

      // Find og kør vores SimpleTool
      const tool = tools.get(toolName);
      if (tool) {
        const result = tool.execute(toolArgs); // <-- DIN kode kører her!
        console.log(`   📤 Result: ${result.substring(0, 80)}...`);

        // 4. Tilføj resultat til messages
        messages.push({ role: "tool", content: result });
      } else {
        messages.push({ role: "tool", content: `Error: Unknown tool ${toolName}` });
      }
    }
    // Loop fortsætter - LLM ser resultaterne
  }

  return "Max iterations reached";
}

console.log("✅ runAgent function defined (~50 lines)");
````

## 1.6 Øvelse 2: Human-in-the-Loop Confirmation
*~15 minutter*

I produktion vil du ofte have tools der gør "farlige" ting - f.eks. starter en ladesession, foretager en betaling, eller sletter data. Før sådanne tools eksekveres, bør agenten **bede om bekræftelse**.

**Scenarie:** Forestil dig at `calculate_charging_cost` kunne udløse en forhåndsreservation eller pre-authorization på dit kort. Før vi eksekverer det, vil vi gerne have brugerens godkendelse.

**Din opgave:** Modificer loopet så det beder om bekræftelse før "farlige" tools eksekveres.

````typescript
// 🎯 ØVELSE 2: Tilføj human-in-the-loop confirmation

// Definer hvilke tools der kræver bekræftelse
const dangerousTools = new Set(["calculate_charging_cost"]);

// RunAgentWithConfirmation - næsten identisk med RunAgent!
// Bruger prompt() fra Deno til at spørge brugeren
async function runAgentWithConfirmation(
  userMessage: string,
  tools: Map<string, SimpleTool>,
): Promise<string> {
  const messages: { role: string; content: string }[] = [
    {
      role: "system",
      content: `Du er en hjælpsom EV-ladnings assistent for ChargeSmart netværket i København.
        Brug de tilgængelige tools til at besvare spørgsmål om ladestationer,
        tilgængelighed, og priser. Svar altid på dansk.`,
    },
    { role: "user", content: userMessage },
  ];

  const ollamaTools = [...tools.values()].map(toOllamaFormat);
  const maxIterations = 10;

  for (let iteration = 1; iteration <= maxIterations; iteration++) {
    console.log(`\n--- Iteration ${iteration} ---`);

    const response = await ollama.chat({
      model: selectedModelName,
      messages,
      tools: ollamaTools,
    });

    messages.push(response.message as any);

    if (!response.message.tool_calls?.length) {
      console.log("✅ LLM finished (no tool calls)");
      return response.message.content || "No response";
    }

    console.log(`🔧 LLM wants to call ${response.message.tool_calls.length} tool(s)`);

    for (const toolCall of response.message.tool_calls) {
      const toolName = toolCall.function.name;
      const toolArgs = toolCall.function.arguments as Record<string, unknown>;
      console.log(`   Tool: ${toolName}`);
      console.log(`   Args: ${JSON.stringify(toolArgs)}`);

      // 🎯 ØVELSE: Bekræftelseslogik!
      if (dangerousTools.has(toolName)) {
        const confirmation = prompt(`⚠️ Execute '${toolName}'? (y/n):`);
        if (confirmation?.trim().toLowerCase() !== "y") {
          console.log("   ⏭️ Skipped by user");
          messages.push({ role: "tool", content: `Tool ${toolName} was skipped by user` });
          continue;
        }
        console.log("   ✅ Confirmed by user");
      }

      const tool = tools.get(toolName);
      if (tool) {
        const result = tool.execute(toolArgs);
        console.log(`   📤 Result: ${result.substring(0, 80)}...`);
        messages.push({ role: "tool", content: result });
      } else {
        messages.push({ role: "tool", content: `Error: Unknown tool ${toolName}` });
      }
    }
  }

  return "Max iterations reached";
}
````

````typescript
// Tool registry: map fra tool-navn til definition + handler
interface ToolDef {
  definition: ReturnType<typeof toOllamaFormat>;
  handler: (args: Record<string, unknown>) => string;
}

const autoTools: Map<string, ToolDef> = new Map([
  ["get_station_status", {
    definition: {
      type: "function" as const,
      function: {
        name: "get_station_status",
        description: "Get real-time status of an EV charging station. Parameter: stationId (e.g. CPH-001)",
        parameters: {
          type: "object",
          properties: { stationId: { type: "string", description: "Station ID like CPH-001" } },
          required: ["stationId"],
        },
      },
    },
    handler: (args) => getStationStatus(args.stationId as string),
  }],
  ["list_stations", {
    definition: {
      type: "function" as const,
      function: {
        name: "list_stations",
        description: "List all EV charging stations in the ChargeSmart network with their current availability",
        parameters: { type: "object", properties: {} },
      },
    },
    handler: (_args) => listStations(),
  }],
  ["calculate_charging_cost", {
    definition: {
      type: "function" as const,
      function: {
        name: "calculate_charging_cost",
        description: "Calculate the cost of charging based on kWh needed and hour of day (0-23)",
        parameters: {
          type: "object",
          properties: {
            kwh: { type: "number", description: "Energy needed in kWh" },
            hour: { type: "integer", description: "Hour of day (0-23)" },
          },
          required: ["kwh", "hour"],
        },
      },
    },
    handler: (args) => calculateChargingCost(Number(args.kwh), Number(args.hour)),
  }],
]);

async function runAgentAuto(userMessage: string): Promise<string> {
  const messages: { role: string; content: string }[] = [
    {
      role: "system",
      content: `Du er en hjælpsom EV-ladnings assistent for ChargeSmart netværket i København.
        Brug de tilgængelige tools til at besvare spørgsmål om ladestationer,
        tilgængelighed, og priser. Svar altid på dansk.`,
    },
    { role: "user", content: userMessage },
  ];

  const ollamaToolDefs = [...autoTools.values()].map(t => t.definition);
  const maxIterations = 10;

  for (let iteration = 1; iteration <= maxIterations; iteration++) {
    console.log(`\n--- Iteration ${iteration} ---`);

    const response = await ollama.chat({
      model: selectedModelName,
      messages,
      tools: ollamaToolDefs,
    });

    messages.push(response.message as any);

    if (!response.message.tool_calls?.length) {
      console.log("✅ LLM finished (no tool calls)");
      return response.message.content || "No response";
    }

    for (const toolCall of response.message.tool_calls) {
      const toolName = toolCall.function.name;
      const toolArgs = toolCall.function.arguments as Record<string, unknown>;
      console.log(`   🔧 Auto-calling: ${toolName}(${JSON.stringify(toolArgs)})`);

      const toolDef = autoTools.get(toolName);
      if (toolDef) {
        const result = toolDef.handler(toolArgs);
        console.log(`   📤 Result: ${result.substring(0, 80)}...`);
        messages.push({ role: "tool", content: result });
      } else {
        messages.push({ role: "tool", content: `Error: Unknown tool ${toolName}` });
      }
    }
  }

  return "Max iterations reached";
}

console.log("✅ runAgentAuto function defined");
console.log("   - Automatisk tool invocation via typed funktioner");
console.log("   - Looper indtil færdig");
````

Kode er ikke bare en use case – det er den universelle grænseflade til den digitale verden.

En coding agent kan:
- Kalde API'er for at hente data (research)
- Organisere data i filsystemet (dokumenter)
- Analysere med Python (finans)
- Generere output i ethvert format (marketing)

**Konklusion**: Vi behøver ikke mange forskellige agents – vi behøver ÉN general-purpose coding agent + **skills** der giver domæneekspertise.

## Teori: Hvorfor Skills?

**Problemet:** Giv en model 100 tools, og den bliver forvirret. Tool selection degraderer.

**Løsningen:** Skills = "lazy-loaded expertise"
- En "router" klassificerer først brugerens intent
- Kun relevante tools/knowledge indlæses
- Modellen fokuserer på én opgave ad gangen

**Key insight:** Copilot har allerede loopet, LLM, og basale tools. Du tilføjer domæneviden!
